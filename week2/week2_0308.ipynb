```python
#7.近四年北中東南師生比？
```

```python
df109t = pd.read_csv('109-test.csv')
df109s = pd.read_csv('109-student.csv')
df108t = pd.read_csv('108-test.csv')
df108s = pd.read_csv('108-student.csv')
df107t = pd.read_csv('107-test.csv')
df107s = pd.read_csv('107-student.csv')
```

```python
df109st = pd.merge(df109t, df109s)
df108st = pd.merge(df108t, df108s)
df107st = pd.merge(df107t, df107s)
```

```python
df109st['教師總人數'] = df109st.iloc[:,3:6].sum(axis=1)
df109st['學生總人數'] = df109st.iloc[:,6:10].sum(axis=1)
df108st['教師總人數'] = df108st.iloc[:,3:6].sum(axis=1)
df108st['學生總人數'] = df108st.iloc[:,6:10].sum(axis=1)
df107st['教師總人數'] = df107st.iloc[:,3:6].sum(axis=1)
df107st['學生總人數'] = df107st.iloc[:,6:10].sum(axis=1)
```

```python
df109area = df109st[['區域', '教師總人數','學生總人數']]
df109asum = df109area.groupby(['區域']).sum().reset_index()
df109asum['師生比'] = df109asum['學生總人數']/df109asum['教師總人數']
df109asum['year'] = 2020
df108area = df108st[['區域', '教師總人數','學生總人數']]
df108asum = df108area.groupby(['區域']).sum().reset_index()
df108asum['師生比'] = df108asum['學生總人數']/df108asum['教師總人數']
df108asum['year'] = 2019
df107area = df107st[['區域', '教師總人數','學生總人數']]
df107asum = df107area.groupby(['區域']).sum().reset_index()
df107asum['師生比'] = df107asum['學生總人數']/df107asum['教師總人數']
df107asum['year'] = 2018
```

```python
frames = [df110asum, df109asum, df108asum, df107asum]
result = pd.concat(frames)
result
```

```python
#8.依照區域排序近四年資料
```
